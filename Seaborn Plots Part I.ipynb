```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import skewnorm
```

```python
#Line Chart using Seaborn Library
x=[10,20,30,40]
y=[0,10,15,25]
sns.set()
plt.plot(x,y)
plt.show()
```

```python
cars=sns.load_dataset('mpg').dropna()
sns.relplot(x='model_year', y='mpg',col='origin',hue='cylinders',data=cars)
```

```python
#univariate KDE Plot
sns.kdeplot(cars.horsepower,shade=True)
```

```python
#bivariate KDE Plots
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower, y=cars.mpg,
            n_levels=15,
            shade=True,
            shade_lowest=False,
            color='red',
            cbar=True)
```

```python
#distribution plot: Histogram, KDE Plot and Rug Plot

sns.distplot(cars.weight,rug=True,bins=15,kde_kws={'lw':3}, rug_kws={'height':0.15})
```

```python
#Distplot with normalised KDE Plots
sns.distplot(cars.weight,bins=8,fit=norm,color='green', kde=False,hist_kws={'alpha':1})
```

```python
#Box Plot

cars=cars[cars.cylinders.isin([4,6,8])]
sns.boxplot(x='origin', y='mpg', data=cars, color='orange')
```

```python
#Comparing boxplots based on no. of cylinders
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars)
```

```python
cars['newer_models']=cars['model_year'] > 76
sns.boxplot(x='origin', y='mpg', hue='newer_models', data=cars, order=['japan', 'europe', 'usa'])
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['europe', 'japan'])],
               split=True, inner='quartiles')
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds=diamonds[diamonds.cut.isin(['Premium', 'Good']) & diamonds.color.isin(['D', 'F', 'J'])].sample(n=100)
sns.set_style('dark')
sns.scatterplot(x='carat',y='price', data=diamonds, hue='depth',s=70)
```
