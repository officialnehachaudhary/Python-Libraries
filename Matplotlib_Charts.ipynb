```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#line chart
plt.figure(figsize=(4,4))
plt.plot([1,2,3,4],[1,4,9,16],marker='o')
plt.axis([0,6,0,20])
plt.title("Square Plot")
plt.xlabel("Number n")
plt.ylabel("n^2")
plt.annotate("Square it", (3,6))
plt.show()
```

```python
#bar chart
x=np.arange(4)
y=[8.4,3.2,1.1,7.5]
plt.xticks(x,('Ankit', 'Hans', 'Joe', 'Neha'))
plt.bar(x,y, color=['orange','brown','blue','green'])
plt.show()
```

```python
#scatter plot
d = {'Red O': np.random.rand(10), 'Blue X': np.random.rand(10)}
df=pd.DataFrame(d)
df.plot(style=['ro','bx'])
plt.show()
```

```python
#time series analysis
ts=pd.Series(np.random.randn(180),index=pd.date_range('1/1/2024',periods=180))
df=pd.DataFrame(np.random.randn(180,3),index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

```python
#Scatter Plot
N=70
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2
plt.scatter(x,y, s=sizes, c=colors, alpha=0.8)
plt.show()
```

```python
#subplot containing log, exp, sin and cos functions
fig=plt.figure()
fig.add_subplot(221,facecolor='aqua')
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222,facecolor='yellow')
plt.plot([np.exp(n) for n in range(1,10)])
fig.add_subplot(223,facecolor='green')
plt.plot([np.sin(n) for n in range(0,20)])
fig.add_subplot(224,facecolor='red')
plt.plot([np.cos(n) for n in range(0,20)])
plt.show()
```

```python
#subplots with a common x-axis
fig, plots=plt.subplots(2,sharex=True)
x=range(0,200,5)
y=[n**2 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y,color='b')
plt.show()
```

```python
#Stem Plot
x=np.linspace(0.1,2*np.pi,40)
y=np.exp(np.sin(x))
plt.stem(x,y)
plt.title("Stem Plot")
plt.show()
```

```python
#Histogram
x = [1,2,3,3,3,6,7,8,9,8]
plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7, 8, 9])
plt.title("Histogram")
plt.show()
```

```python
#Stack Plot
days=['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
studying=[7,8,4,2,1,4,10]
playing=[1,2,1,2,3,1,1]
plt.stackplot(days,studying, playing, colors=['r','c'])
plt.title('Hours spend on studying and playing wrt to Days',color='blue')
plt.legend(['Study','Play'])
plt.show()
```

```python
#Box Plot
data1=np.random.normal(100,20,200)
data2=np.random.normal(50,70,100)
data3=np.random.normal(200,20,100)
dataSet=[data1,data2,data3]
plt.figure(figsize=(5,3))
plt.boxplot(dataSet)
plt.show()
```

```python
#Pie Chart with explosion
x={'Products':['A','B','C','D','E'],'values':[4,5,1,3,4]}
df=pd.DataFrame(x)
e=[0.1,0,0,0,0]
plt.pie(df['values'],labels=df['Products'], autopct='%1.1f%%', explode=e)
plt.show()
```

```python
#Error plot
x=[1,2,3,4,5,6,7]
y=[5,2,4,2,5,3,1]

y_error=0.2

plt.plot(x,y)
plt.errorbar(x,y,yerr=y_error,marker='o')
plt.show()
```
